```python
def performCode(x):
    strList = x.split()
    newStr = f'{strList[0]} {strList[1]}'
    strList.append(newStr)
    return strList[2]
```

```python
def performNumber(x):
    strList = x.split()
    newStr = f'{strList[0]} {strList[1]}'
    strList.append(newStr)
    return strList[3]
```

```python
def performLibrary(x):
    strList = x.split()
    newStr = f'{strList[0]} {strList[1]}'
    strList.append(newStr)
    return strList[4]
```

```python
df['Code'] = df['Shelfmarks'].map(performCode)
df['Number'] = df['Shelfmarks'].map(performNumber)
df['Library'] = df['Shelfmarks'].map(performLibrary)
```
